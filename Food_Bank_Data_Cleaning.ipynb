## Removes Percentage Sign '%' and Changes Type from str -> float

```python
def remove_percent(df):
    new = []
    for i in df:
        if i == None or isinstance(i, float):
            new.append(math.nan)
        else:
            i = i.replace('%','')
            new.append(float(i))
    return new
```

## Removes the commas and Money signs $ and changes type from str -> float

```python
def remove_money_sign(df):
    new = []
    for i in df:
        if i == None or isinstance(i, float):
            new.append(math.nan)
        else:
            i = i.replace('$','')
            i = i.replace(',','')
            new.append(float(i))
    return new
```
